# FETCHING TEXT FROM TARGETED DIV TAG

```python
def texts():
    global text
    text = []
    for links in hold_urls:
        rsp = urls(links)
        content = BeautifulSoup(response.text,'lxml')
        main_content = content.find_all('div',class_ = "a52608cc")
        for ins in main_content:
            text.append(ins.text)
texts()
```

# DATA READING

```python
for index,txt in enumerate(text):
    text[index] = txt.upper()
text
```

# DATA PREPROCESSING

```python
text[0]
```

```python
pattern = '\d[a-zA-z]+'
re.search(pattern,text[0]).group()
```

# DATA ANALYSING AND FILTERING AND CLEANING

```python
all_words = []
car = []
hold_temp = []
all_brands = []
for indexes,txt in enumerate(text):
    pattern = '\d[a-zA-z]+'
    brand = re.search(pattern,txt).group()
    start_index = re.search(pattern,txt).end()
    words = txt[start_index:]
    pattern = r'\w+'
    word = re.findall(pattern, words)
    if 'KM' in word:
        all_words.append(word)
        for index,km in enumerate(word):
            if km == 'KM':
                start = index-2
                temp = word[start] + word[start+1] + word[index]
                car.append(word[0])
                hold_temp.append(temp)
                all_brands.append(brand[1:])
```
